```python
# 不要組み合わせの除外

# 初期化
df['del_flg'] = 0

# ①SweaterAndShirt（CLOTHING）× Bandana（HEAD）
df.loc[(df['CLOTHING']=='SweaterAndShirt') & (df['HEAD']=='Bandana'), 'del_flg'] = 1

# ②Jacket（CLOTHING）× GreenCap（HEAD）
df.loc[(df['CLOTHING']=='Jacket') & (df['HEAD']=='GreenCap'), 'del_flg'] = 1

# ③Jacket（CLOTHING）× Bandana（HEAD）
df.loc[(df['CLOTHING']=='Jacket') & (df['HEAD']=='Bandana'), 'del_flg'] = 1

# ④BigHoodie（HEAD）× 全てのCLOTIHNGパーツ
# ※プログラム的に組み合わせる必要があるのであれば、以下のCLOTHINGパーツがはみ出てしまう為NG
# KnitSweater、T-shirt、GrHoodieLayered、BlackZip-Up、WhHoodieLayered
df.loc[(df['CLOTHING']!='BlackCut_Sewn') & (df['HEAD']=='BigHoodie'), 'del_flg'] = 1

df.loc[(df['CLOTHING']=='BlackCut_Sewn') & (df['HEAD']=='BigHoodie'), 'Properties'] = df.loc[(df['CLOTHING']=='BlackCut_Sewn') & (df['HEAD']=='BigHoodie'), 'Properties'].str.replace("BlackCut-Sewn", "None")

# ⑤GasMask（HEAD）× WhRimmedGlasses（FACE）
df.loc[(df['HEAD']=='GasMask') & (df['FACE']=='WhRimmedGlasses') , 'del_flg'] = 1

# ⑥GasMask（HEAD）× BrTintedGlasses（FACE）
df.loc[(df['HEAD']=='GasMask') & (df['FACE']=='BrTintedGlasses') , 'del_flg'] = 1

# ⑦GasMask（HEAD）× BkShadedGlasses（FACE）
df.loc[(df['HEAD']=='GasMask') & (df['FACE']=='BkShadedGlasses') , 'del_flg'] = 1

# ⑧GasMask（HEAD）× BkRoundGlasses（FACE）
df.loc[(df['HEAD']=='GasMask') & (df['FACE']=='BkRoundGlasses') , 'del_flg'] = 1

```

```python
# データフレームの読み込み
df = pd.read_csv('list_ホクロ追加.csv')

# 移動先のフォルダを指定
destination_folder = './hokuro/'

# 再帰的に探索するフォルダを指定
search_folder = './納品物/'

# 再帰的に探索する
for root, dirs, files in os.walk(search_folder):
    display(len(files))
    for file in files:
        # ファイル名を取得
        file_name = os.path.basename(file)
        
        # データフレーム内にファイル名が存在する場合は移動
        if file_name in df['Media file name'].tolist():
            # 移動先のファイルパスを生成
            destination_file_path = os.path.join(destination_folder, file_name)
            
            # ファイルを移動
            shutil.move(os.path.join(root, file), destination_file_path)
        
            
```

```python
#フォルダの個数をカウント（listdirの引数はファイルのあるフォルダのパス）
num = len(os.listdir('./hashlips_art_engine-main/build/json/'))

#読み込むファイルのパスを指定
for i in range(0, num):
    tmp_df = pd.read_json('./hashlips_art_engine-main/build/json/' + str(i+1) + '.json')
    
    for j in tmp_df.index:
        dic = tmp_df.at[j,"attributes"]
        tmp_df.at[0, dic["trait_type"]] = dic["value"]
    tmp_df.at[0, "image"] =  os.path.basename(tmp_df.loc[0, "image"])
    tmp_df.at[0, "Media type"] = "IMAGE"

    if i == 0:
        df = tmp_df.head(1)
    else:
        df = pd.concat([df, tmp_df.head(1)])
    
# indexを再採番    
df.index = np.arange(1, len(df)+1)

df['new_file_name'] = df['img'].str.cat(['.png']*len(df))

# # Propertiesの生成
# for i, row in df.iterrows():
#     df.loc[i, "Properties"] = df.loc[i,['BACKGROUND', 'FACE', 'CLOTHING', 'HEAD']].to_json()
# df['Properties'] = df['Properties'].str.replace(":", ": ")
# df['Properties'] = df['Properties'].str.replace(",", ", ")
# df['Properties'] = df['Properties'].str.replace("_", "-")

# # 不要列削除
# df = df.drop(["description", "dna", "compiler", "date", "attributes"], axis=1)

# # LINE NFT 用 カラム名リネーム
# df = df.rename(columns={'edition': 'Number', 'name': 'Token Name', "image": "Media file name"})
# df = df.reindex(columns=['Number', 'Token Name', 'Properties', 'Media file name', "Media type", 'BACKGROUND', 'FACE', 'CLOTHING', 'HEAD'])
# df['Token Name'] = 'JIN TEN'

df
```

```python
# dfの'image'を'new_file_name'にリネームして移動する

# ファイルをリネームして移動する関数の定義
def rename_and_copy(file_name, new_name):
    shutil.copy("./hashlips_art_engine-main/build/images/"+file_name, "./ホクロ追加/"+new_name)

# DataFrameにapply関数を使って、ファイルをリネームして移動する
df.apply(lambda x: rename_and_copy(x['image'], x['new_file_name']), axis=1)
```
